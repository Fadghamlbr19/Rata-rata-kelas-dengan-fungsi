```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import pandas as pd
```

```python
image = cv.imread("/content/drive/MyDrive/white.jpeg")
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Define the step size for the second derivative
h = 1

# Compute the second derivative of the image using the Taylor series formula
dxx = (gray[:-2, 1:-1] - 2 * gray[1:-1, 1:-1] + gray[2:, 1:-1]) / (h ** 2)
dyy = (gray[1:-1, :-2] - 2 * gray[1:-1, 1:-1] + gray[1:-1, 2:]) / (h ** 2)
dxy = (gray[:-2, :-2] - gray[:-2, 2:] - gray[2:, :-2] + gray[2:, 2:]) / (4 * h ** 2)
```

```python
# Compute the Laplacian of the image
laplacian = dxx + dyy

# Normalize the Laplacian to the range [0, 255]
laplacian = cv.normalize(laplacian, None, 0, 255, cv.NORM_MINMAX, cv.CV_8U)

# Threshold the Laplacian to identify edges
thresh = 100
edges = cv.threshold(laplacian, thresh, 255, cv.THRESH_BINARY)[1]
```

```python
# Display the edges
plt.imshow(edges,cmap='gray')
plt.show()
```
